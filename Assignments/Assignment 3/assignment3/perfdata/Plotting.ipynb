```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Manually entering the data
data = {
    "number of threads": [1, 2, 4],
    "Basic": [1, 1, 1],  # Example values, replace with your data
    "Reduced Default Schedule": [1.74, 1.18, 1.09],  # Replace with your data
    "Reduced Forces Cyclic": [1.72, 1.70, 1.70],  # Replace with your data
    "Reduced All Cyclic": [1.73, 1.68, 1.48]  # Replace with your data
}
```

```python
df = pd.DataFrame(data)

# Set the number of threads as the index
df.set_index("number of threads", inplace=True)
```

```python
# Plot the data
plt.figure(figsize=(8, 6))
plt.plot(df.index, df["Basic"], marker='o', label="Basic")
plt.plot(df.index, df["Reduced Default Schedule"], marker='o', label="Reduced Default Schedule")
plt.plot(df.index, df["Reduced Forces Cyclic"], marker='o', label="Reduced Forces Cyclic")
plt.plot(df.index, df["Reduced All Cyclic"], marker='o', label="Reduced All Cyclic")

# Customize the plot
plt.title("Efficiencies with Respect to 'Basic'")
plt.xlabel("Threads")
plt.ylabel("Efficiency")
plt.legend(bbox_to_anchor=(1.05, 0.5), loc='center left', borderaxespad=0.)
plt.grid(True)

# Show the plot
plt.show()
```
